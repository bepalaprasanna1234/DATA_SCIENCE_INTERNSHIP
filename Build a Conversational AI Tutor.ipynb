```python
import random

class MathTutor:
    def __init__(self):
        self.topics = ["addition", "subtraction", "multiplication", "division"]
    
    def start(self):
        print("Hi there! I'm your Math Tutor. What would you like to learn today?")
        print("Here are the topics I can help you with:")
        for topic in self.topics:
            print("- " + topic)
        topic_choice = input("Please choose a topic: ").lower()
        if topic_choice in self.topics:
            self.learn(topic_choice)
        else:
            print("Sorry, I can't help with that topic right now.")
    
    def learn(self, topic):
        print("Great choice! Let's learn about", topic)
        if topic == "addition":
            self.learn_addition()
        elif topic == "subtraction":
            self.learn_subtraction()
        elif topic == "multiplication":
            self.learn_multiplication()
        elif topic == "division":
            self.learn_division()
    
    def learn_addition(self):
        print("Addition is combining two or more numbers to find their total sum.")
        num1 = random.randint(1, 10)
        num2 = random.randint(1, 10)
        answer = num1 + num2
        user_answer = input(f"What is {num1} + {num2}? ")
        if int(user_answer) == answer:
            print("Correct! Well done!")
        else:
            print("Incorrect. The correct answer is", answer)
    
    def learn_subtraction(self):
        print("Subtraction is taking one number away from another to find the difference.")
        num1 = random.randint(1, 10)
        num2 = random.randint(1, num1)
        answer = num1 - num2
        user_answer = input(f"What is {num1} - {num2}? ")
        if int(user_answer) == answer:
            print("Correct! Well done!")
        else:
            print("Incorrect. The correct answer is", answer)
    
    def learn_multiplication(self):
        print("Multiplication is repeated addition.")
        num1 = random.randint(1, 10)
        num2 = random.randint(1, 10)
        answer = num1 * num2
        user_answer = input(f"What is {num1} * {num2}? ")
        if int(user_answer) == answer:
            print("Correct! Well done!")
        else:
            print("Incorrect. The correct answer is", answer)
    
    def learn_division(self):
        print("Division is splitting a number into equal parts.")
        dividend = random.randint(1, 10)
        divisor = random.randint(1, dividend)
        answer = dividend // divisor
        user_answer = input(f"What is {dividend} รท {divisor}? ")
        if int(user_answer) == answer:
            print("Correct! Well done!")
        else:
            print("Incorrect. The correct answer is", answer)

# Usage
tutor = MathTutor()
tutor.start()
```
